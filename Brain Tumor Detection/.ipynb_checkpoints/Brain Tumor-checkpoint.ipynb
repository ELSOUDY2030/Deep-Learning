dataset from https://www.kaggle.com/datasets/masoudnickparvar/brain-tumor-mri-dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import pandas as pd
import matplotlib.colors as Colormap
import seaborn as sns
import os

from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import tensorflow as tf
import keras
from keras import layers
from keras import utils

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data_dir = 'Training\\'
width = 180
batch_size = 32
train_ds = tf.keras.utils.image_dataset_from_directory(data_dir,
                                                       validation_split=0.2,
                                                       subset="training",
                                                       seed=42,
                                                       image_size=(width,width),
                                                       batch_size=batch_size


)
```

```python
data_dir = 'Training\\'
width = 180
batch_size = 32
val_ds = tf.keras.utils.image_dataset_from_directory(data_dir,
                                                       validation_split=0.2,
                                                       subset="validation",
                                                       seed=42,
                                                       image_size=(width,width),
                                                       batch_size=batch_size


)
```

```python
# Get labels inferred by the method itself
class_names = train_ds.class_names
print(class_names)
```

```python
train_ds_labels = []
for _,labels in train_ds:
    for i in range(len(labels)):
        train_ds_labels.append(class_names[labels[i].numpy()])

        

      
```

```python
# Create a function to customize autopct parameter of plt.pie()
def make_autopct(values):
    def my_autopct(pct):
        # The pct is percentage value that matplotlib supplies for every wedge
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.2f}%  ({val})'
    return my_autopct
```

```python
df = pd.DataFrame({'Category':train_ds_labels})
```

```python
cat = df.Category.value_counts().index.to_list()
cat = [i.title() for i in cat]
count = df.Category.value_counts().to_list()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,8))
plt.pie(count,labels=cat,shadow=True,autopct=make_autopct(count))
plt.title('Distribution')
plt.show()
```

```python
plt.figure(figsize=(15,8))
for image, labels in train_ds.take(23):
    for i in range(15):
        ax = plt.subplot(3,5,1+i)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.axis('off')
        plt.title(class_names[labels[i]],fontsize=12,pad=8)
plt.subplots_adjust(wspace=0.25, hspace=0.25)
plt.show()
```

```python
Autotune = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=Autotune)
val_ds = val_ds.cache().prefetch(buffer_size=Autotune)
```

```python
Normalization_layers = layers.Rescaling(1/255)

data_augmentation = keras.Sequential([
    layers.RandomFlip('horizontal',
                     input_shape=(width,width,3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])
```

```python
def plot_images_augmented(dataset):
    """
    Parameters:
    : dataset - dataset of type `tf.data.Dataset` 
                with images to plot
    """
    plt.figure(figsize=(10, 10))
    for images, _ in dataset.take(1):
        for i in range(9):
            augmented_images = data_augmentation(images)
            ax = plt.subplot(3, 3, i + 1)
            plt.imshow(augmented_images[0].numpy().astype("uint8"))
            plt.axis("off")
```

```python
plot_images_augmented(train_ds)
```

```python
num_classes = len(class_names)
model = keras.Sequential([
    data_augmentation,
    Normalization_layers,
    layers.Conv2D(32,3,padding='same',activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),
    
    layers.Conv2D(64,3,padding='same',activation='relu'),    
    layers.MaxPooling2D(pool_size=(2,2)),
    
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),    
    
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)), 
    
    layers.Dropout(0.3),
    layers.Flatten(),
    
    layers.Dense(265,activation='relu'),
    layers.Dense(num_classes,activation='softmax' ,name='output')
   
    
])
```

```python
model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
checkpoint = keras.callbacks.ModelCheckpoint('brainT_detect.h5')

early_stop = keras.callbacks.EarlyStopping(patience=10,
                                          restore_best_weights=True)
```

```python
his_model = model.fit(train_ds,
                     validation_data=val_ds,
                     epochs=100,
                     callbacks=[checkpoint,early_stop])
```

```python
epochs = len(his_model.epoch) # epochs completed by early stopping
accuracy = his_model.history['accuracy']
val_acc = his_model.history['val_accuracy']

loss = his_model.history['loss']
val_loss = his_model.history['val_loss']

epochs_range = range(epochs)

# Plot training and validation accuracy
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, accuracy, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

# Plot training and validation loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
model.save("brainT_detect.h5")
```
